```python
def calculate_total_price(price_per_item, quantity):
    """
    Calculates the total price of items.

    Args:
        price_per_item (float): The price of a single item.
        quantity (int): The number of items purchased.

    Returns:
        float: The total calculated price.
    """
    total_price = price_per_item * quantity
    return total_price
```

```python
try:
    item_price = float(input("Enter the price of one item: "))
    number_of_items = int(input("Enter the number of items purchased: "))

    # Calculate the total price using the function
    final_total = calculate_total_price(item_price, number_of_items)

    # Display the total price
    print(f"\nPrice per item: ${item_price:.2f}")
    print(f"Quantity: {number_of_items}")
    print(f"Total price: ${final_total:.2f}")
except ValueError:
    print("Invalid input. Please enter valid numbers for price and quantity.")
```
