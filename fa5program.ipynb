```python
def calculate_total(price, quantity):
    """Calculates total price based on price and quantity."""
    total = price * quantity
    return total

print("--- Personal Item Calculator ---")
try:
    user_price = float(input("Enter the price of the item: "))
    user_quantity = int(input("Enter the number of items purchased: "))

    # Step 3: Call the function and display the result
    grand_total = calculate_total(user_price, user_quantity)
    print(f"\nTotal Price: ${grand_total:.2f}")

except ValueError:
    print("Invalid input. Please enter numbers only.")
```
